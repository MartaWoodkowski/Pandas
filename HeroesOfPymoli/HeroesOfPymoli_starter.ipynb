```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(15)
```

## Player Count

* Display the total number of players


```python
Players = purchase_data['SN'].unique()
Total_Players = len(Players)

#Created the Data Frame to show Total Players
Number_of_Players_df = pd.DataFrame({'Total Players':[Total_Players]})
Number_of_Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items = purchase_data['Item ID'].unique()
Number_of_Unique_Items = len(Unique_Items)
Average_Price = purchase_data['Price'].mean()
Number_of_Purchases = purchase_data['Purchase ID'].count()
Total_Revenue = purchase_data['Price'].sum()

#Created Data Frame to show my analysis wich contains the calculations above
Purchasing_Analysis_df = pd.DataFrame({'Number of Unique Items':[Number_of_Unique_Items],
                                      'Average Price':[Average_Price],
                                      'Number of Purchases':[Number_of_Purchases],
                                      'Total Revenue':[Total_Revenue]})

#Formated 'Average Price' and 'Total Revenue' columns to show dollar sign + to show two decimal points 
Purchasing_Analysis_df['Average Price'] = Purchasing_Analysis_df['Average Price'].astype(float).map(
    "${:,.2f}".format)
Purchasing_Analysis_df['Total Revenue'] = Purchasing_Analysis_df['Total Revenue'].astype(float).map(
    "${:,.2f}".format)

Purchasing_Analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Created Unique Players DataFrame using drop_duplicates function
Unique_Players_df = purchase_data[['SN','Gender']].drop_duplicates()

#Counted Male Players
Male_Players = Unique_Players_df.loc[Unique_Players_df['Gender'] == 'Male', :]
Count_Male = len(Male_Players)
Percent_Male_Players = Count_Male/Total_Players

#Counted Female Players
Female_Players = Unique_Players_df.loc[Unique_Players_df['Gender'] == 'Female', :]
Count_Female = len(Female_Players)
Percent_Female_Players = Count_Female/Total_Players

#Counted Non-Disclosed Players
NonDisclosed_Players = Unique_Players_df.loc[Unique_Players_df['Gender'] == 'Other / Non-Disclosed', :]
Count_NonDisclosed = len(NonDisclosed_Players)
Percent_NonDisclosed_Players = Count_NonDisclosed/Total_Players


#Created Gender Demographics Data Frame
Gender_Demographics_df = pd.DataFrame({'Total Count':[Count_Male,Count_Female,Count_NonDisclosed],
                                       'Percentage of Players':[Percent_Male_Players,Percent_Female_Players,Percent_NonDisclosed_Players]
                                      })
Gender_Demographics_df = Gender_Demographics_df.set_axis(['Male','Female','Other / Non-Disclosed'])

#Formated 
Gender_Demographics_df['Percentage of Players'] = Gender_Demographics_df['Percentage of Players'].astype(float).map('{:.2%}'.format)

Gender_Demographics_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchasing_Analysis_df = purchase_data[['Purchase ID', 'Gender', 'Price']]

#Grouped by Gender. Calculated and renamed 'Purchase Count' column 
Purchasing_Analysis = Purchasing_Analysis_df.groupby(['Gender'])
Analysis_by_Gender = Purchasing_Analysis.count()
Analysis_by_Gender = Analysis_by_Gender.rename(columns = {'Purchase ID':'Purchase Count'})

#Calculated, added & formated 'Average Purchase Price' column 
Avg_Price = Purchasing_Analysis['Price'].mean()
Analysis_by_Gender['Average Purchase Price'] = Avg_Price
Analysis_by_Gender['Average Purchase Price'] = Analysis_by_Gender['Average Purchase Price'].astype(float).map(
    "${:,.2f}".format)

#Calculated,added and formated 'Total Purchase Value' column
Purchase_Value_per_Person = Purchasing_Analysis['Price'].sum()
Analysis_by_Gender['Total Purchase Value'] = Purchase_Value_per_Person
Analysis_by_Gender['Total Purchase Value'] = Analysis_by_Gender['Total Purchase Value'].astype(float).map(
    "${:,.2f}".format)

#Calculated, added & formated 'Avg Total Purchase per Person' column
Purchase_Value = Purchasing_Analysis['Price'].sum()
Players_by_Gender = Unique_Players_df.groupby(['Gender'])
Player_Count_by_Gender = Players_by_Gender.count()
Avg_Purchase_per_Person = Purchase_Value/Player_Count_by_Gender['SN']
Analysis_by_Gender['Avg Total Purchase per Person'] = Avg_Purchase_per_Person
Analysis_by_Gender['Avg Total Purchase per Person'] = Analysis_by_Gender['Avg Total Purchase per Person'].astype(float).map(
    "${:,.2f}".format)

#Deleted column 'Price' (I didn't need to keep it) and removed column name ('Gender')
del Analysis_by_Gender ['Price']
Analysis_by_Gender.index.name = None

Analysis_by_Gender.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Figure out the minimum and maximum views for Age in our Data
#print(purchase_data['Age'].max())
#print(purchase_data['Age'].min())

# Create bins 
bins = [0,9,14,19,24,29,34,39,purchase_data['Age'].max()]

# Create labels for these bins
group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Filtered my data to get unique induviduals, and placed it into bins using pd.cut() 
purchase_data_unique = purchase_data[['SN','Age']].drop_duplicates()
purchase_data['Age Group']=pd.cut(purchase_data_unique['Age'], bins, labels=group_labels)
Age_Groups_df = purchase_data[['Age Group','Age','SN']]
Grouped_by_Bins = Age_Groups_df.groupby(['Age Group'])

#Removed my 'Age Group' column name
Grouped_by_Age_df = Grouped_by_Bins.count()
Grouped_by_Age_df.index.name = None

#Calculated, added and formated 'Percentage of Players' column
Percentage_of_Players_by_Age = Grouped_by_Age_df['Age']/Total_Players*100
Grouped_by_Age_df['Percentage of Players'] = Percentage_of_Players_by_Age
Grouped_by_Age_df['Percentage of Players'] = Grouped_by_Age_df['Percentage of Players'].astype(float).map(
    "{:,.2f}%".format)

#Renamed my column from 'Age'to 'Total Count', and deleted the column I didn't need
Grouped_by_Age_df = Grouped_by_Age_df.rename(columns = {'Age':'Total Count'})
del Grouped_by_Age_df['SN']

Grouped_by_Age_df.head(10)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
master_df = purchase_data
master_df['Age Ranges']=pd.cut(purchase_data['Age'], bins, labels=group_labels)

#Created and grouped columns (to my new Data Frame 'Purchasing_Analysis_by_Age_df')
Purchasing_Analysis_by_Age_df = master_df.groupby(['Age Ranges']).count()
Purchasing_Analysis_by_Age_df = Purchasing_Analysis_by_Age_df.rename(columns = {'Purchase ID':'Purchase Count'}) #renamed my first column

Mean_by_Age = master_df.groupby(['Age Ranges']).mean()
Purchasing_Analysis_by_Age_df['Average Purchase Price']=Mean_by_Age['Price']

Total_by_Age = master_df.groupby(['Age Ranges']).sum()
Purchasing_Analysis_by_Age_df['Total Purchase Value']=Total_by_Age['Price']

Purchasing_Analysis_by_Age_df['Avg Total Purchase per Person'] = Total_by_Age['Price']/Grouped_by_Age_df['Total Count']

#Deleted the columns I don't need
Purchasing_Analysis_by_Age_df = Purchasing_Analysis_by_Age_df.drop(['SN','Age','Gender','Item ID','Item Name','Price','Age Group'],axis=1)

#Formated 'Average Purchase Price', 'Total Purchase Value' & 'Avg Total Purchase per Person' columns
Purchasing_Analysis_by_Age_df['Average Purchase Price'] = Purchasing_Analysis_by_Age_df['Average Purchase Price'].astype(float).map(
    "${:,.2f}".format)
Purchasing_Analysis_by_Age_df['Total Purchase Value'] = Purchasing_Analysis_by_Age_df['Total Purchase Value'].astype(float).map(
    "${:,.2f}".format)
Purchasing_Analysis_by_Age_df['Avg Total Purchase per Person'] = Purchasing_Analysis_by_Age_df['Avg Total Purchase per Person'].astype(float).map(
    "${:,.2f}".format)

Purchasing_Analysis_by_Age_df.head(10)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Exctracted the columns I need for this task
Spenders = purchase_data[['SN', 'Purchase ID', 'Price']]

#Grouped and created columns (to my new Data Frame 'Spenders_grouped_df')
Spenders_grouped_df = Spenders.groupby(['SN']).count()
Spenders_grouped_df = Spenders_grouped_df.rename(columns = {'Purchase ID':'Purchase Count'}) #renamed my first column

Mean = Spenders.groupby(['SN']).mean()
Spenders_grouped_df['Average Purchase Price']=Mean['Price']

Total = Spenders.groupby(['SN']).sum()
Spenders_grouped_df['Total Purchase Value']=Total['Price']

#Deleted one of the columns I don't need
del Spenders_grouped_df['Price']

#Sorted in Descending order
Top_Spenders_df = Spenders_grouped_df.sort_values('Total Purchase Value', ascending=False)

#Formated 'Average Purchase Price' & 'Total Purchase Value' columns
Top_Spenders_df['Average Purchase Price'] = Top_Spenders_df['Average Purchase Price'].astype(float).map(
    "${:,.2f}".format)
Top_Spenders_df['Total Purchase Value'] = Top_Spenders_df['Total Purchase Value'].astype(float).map(
    "${:,.2f}".format)

Top_Spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Exctracted the columns I need for this task
Items = purchase_data[['Item ID', 'Item Name','Purchase ID', 'Price']]

#Grouped and created columns (to my new Data Frame 'Item_Prices_df')
Items_grouped_df = Items.groupby(['Item ID','Item Name']).count()
Items_grouped_df = Items_grouped_df.rename(columns = {'Purchase ID':'Purchase Count'}) 

Item_Prices = Items.groupby(['Item ID','Item Name']).mean()
Items_grouped_df['Item Price'] = Item_Prices['Price']

Purchase_Value = Items.groupby(['Item ID','Item Name']).sum()
Items_grouped_df['Total Purchase Value'] = Purchase_Value['Price']

#Deleted one of the columns I don't need
del Items_grouped_df['Price']

#Sorted in Descending order
Top_Items_df = Items_grouped_df.sort_values('Purchase Count', ascending=False)

#Formated 'Item Price' & 'Total Purchase Value' columns
Top_Items_df['Item Price'] = Top_Items_df['Item Price'].astype(float).map(
    "${:,.2f}".format)
Top_Items_df['Total Purchase Value'] = Top_Items_df['Total Purchase Value'].astype(float).map(
    "${:,.2f}".format)
    
Top_Items_df.head()
```
